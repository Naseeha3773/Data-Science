<center>
    <h1>Dropping Constant Features</h1>
</center>

- Using this technique we can remove all the columns that have constant values or very less variance that doesn't help in solving the problem statement.
- In this technique we initially decide a threshold variance value and based on the threshold variance the features are removed.

```python
import numpy as np
import pandas as pd

from sklearn.feature_selection import VarianceThreshold
```

```python
sample_data = pd.DataFrame({
    "A":[1,2,4,1,2,4], 
    "B":[4,5,6,7,8,9], 
    "C":[0,0,0,0,0,0],
    "D":[1,1,1,1,1,1]
})
```

```python
sample_data
```

```python
variance_threshold = VarianceThreshold(threshold=0)
variance_threshold.fit(sample_data)
```

```python
variance_threshold.get_support()
```

```python
columns_to_drop = sample_data.columns[~variance_threshold.get_support()]
columns_to_drop
```

```python
sample_data.drop(columns_to_drop, axis = 1)
```

## Applying Variance Threshold on Large Dataset

Dataset Link: https://www.kaggle.com/competitions/santander-customer-satisfaction/data

```python
data = pd.read_csv("train.csv", nrows=10000)
data.head()
```

```python
data.shape
```

```python
X=data.drop(labels=['TARGET'], axis=1)
y=data['TARGET']
```

```python
variance_threshold = VarianceThreshold(threshold=0)
variance_threshold.fit(X)
```

```python
X.shape
```

```python
sum(variance_threshold.get_support())
```

```python
total_constant_features = X.shape[1] - sum(variance_threshold.get_support())
total_constant_features
```

```python
constant_columns = X.columns[~variance_threshold.get_support()]
constant_columns
```

```python
len(constant_columns)
```

```python
X.drop(columns = constant_columns, axis = 1)
```
